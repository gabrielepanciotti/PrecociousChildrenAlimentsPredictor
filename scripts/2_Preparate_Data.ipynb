```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import parallel_coordinates
from IPython.display import display


# warnings -> to silence warnings
import warnings
warnings.filterwarnings("ignore")
dataset = pd.read_excel('../data/dataset_final.xlsx', engine='openpyxl')
display(dataset)
dataset_filtered = pd.read_excel('../data/dataset_filtered.xlsx', engine='openpyxl')
display(dataset_filtered)
```

```python
var_passed = ['Peso', 'daily_weight_diff', 
              'all_cho','all_prot','all_lip',
              'VolumeNE','ne_calkg','ne_chokg', 'ne_protkg','ne_lipkg',
              'VolumeNP', 'np_calkg', 'np_chokg', 'np_protkg', 'np_lipkg']
columns_removed = ['np_protkg1', 'np_lipkg1', 'np_chokg1', 'np_calkg1',
                    'np_protkg2', 'np_lipkg2', 'np_chokg2', 'np_calkg2',
                    'np_protkg3', 'np_lipkg3', 'np_chokg3', 'np_calkg3',
                    'peso1','peso2','peso3',
                    'ne_calkg1','ne_calkg2','ne_calkg3',
                    'VolumeNP1','VolumeNP2','VolumeNP3'
                    ]

new_dataset = dataset.drop(columns=columns_removed)
display(new_dataset)

for var in var_passed:
    for i in range(1, 11):
        new_dataset[var+f'_d-{i}'] = new_dataset[var].shift(i)
        print(var+f'_d-{i}')

new_dataset.fillna(0, inplace=True)
display(new_dataset)
new_dataset.to_excel('../data/new_dataset.xlsx')
```
