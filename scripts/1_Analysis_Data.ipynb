```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import parallel_coordinates
from IPython.display import display


# warnings -> to silence warnings
import warnings
warnings.filterwarnings("ignore")

dataset = pd.read_excel('../data/new_dataset.xlsx', engine='openpyxl')
display(dataset)
dataset_filtered = pd.read_excel('../data/dataset_filtered.xlsx', engine='openpyxl')
display(dataset_filtered)
```

```python
# 1. Pulizia dei Dati
# Rimozione di eventuali righe con valori mancanti
cleaned_data = dataset.dropna()

# 2. Analisi Esplorativa
# Statistiche di base
basic_stats = cleaned_data.describe()

# 3. Visualizzazione dei Dati
# Grafico della distribuzione del peso
plt.figure(figsize=(10, 6))
sns.histplot(cleaned_data['Peso'], kde=True)
plt.title('Distribuzione del Peso dei Pazienti')
plt.xlabel('Peso (kg)')
plt.ylabel('Frequenza')
plt.grid(True)
plt.show()

# Grafico delle calorie di PN per chilogrammo
plt.figure(figsize=(10, 6))
sns.histplot(cleaned_data['np_calkg'], kde=True, color='orange')
plt.title('Distribuzione delle Calorie di NP per Chilogrammo')
plt.xlabel('Calorie di NP per kg')
plt.ylabel('Frequenza')
plt.grid(True)
plt.show()

# 4. Heatmap delle Correlazioni
# Selezione delle colonne rilevanti per la correlazione
columns_of_interest = ['daily_weight_diff', 'VolumeNP', 'np_calkg', 'np_protkg', 'np_lipkg', 'np_chokg']
correlation_data = cleaned_data[columns_of_interest].corr()

# Visualizzazione della heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_data, annot=True, cmap='coolwarm')
plt.title('Heatmap delle Correlazioni tra Variazione del peso e Nutrizione Parenterale')
plt.show()

columns_of_interest = ['daily_weight_diff', 'VolumeNE', 'ne_calkg', 'ne_protkg', 'ne_lipkg', 'ne_chokg']
correlation_data = cleaned_data[columns_of_interest].corr()

# Visualizzazione della heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_data, annot=True, cmap='coolwarm')
plt.title('Heatmap delle Correlazioni tra Variazione del peso e Nutrizione Enterenale')
plt.show()

# 5. Analisi di Correlazione Avanzata
# Calcolo delle correlazioni tra il peso e le variabili nutrizionali
correlation_results = correlation_data['daily_weight_diff'].sort_values(ascending=False)

# Restituzione dei risultati
basic_stats, correlation_results
```
